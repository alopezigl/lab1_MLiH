a) Write the code to construct from scratch an **unidirected** random graph with 10 nodes using the Gilbert model, i.e., the probability of an edge existing between a pair of nodes is p. For this exercise use p = 0.4.

- *Note that you cannot use any of the functions that automatically creates a random graph from NetworkX library.*

````python
#Solution Milestone 1 question a)

n = 10
p = 0.4

def undirected_gilbert(n, p):
  n = n
  p = p

  # We create an empty undirected graph
  graph = nx.Graph()

  # We add n nodes to the graph
  graph.add_nodes_from(range(n))

  # We iterate over every possible unique pair of nodes
  for i in range(n):
      for j in range(n):
          if i != j:
            if random.random() < p:
                graph.add_edge(i, j)
  return graph

G_gilbert = undirected_gilbert(n, p)
print(f"Custom Gilbert Graph (n={n}, p={p}):")
print(f"Nodes: {G_gilbert.number_of_nodes()}")
print(f"Edges: {G_gilbert.number_of_edges()}")

plt.figure(figsize=(7, 5))
pos = nx.spring_layout(G_gilbert)
nx.draw_networkx(G_gilbert, pos, with_labels=True, node_color='lightblue', node_size=700, font_weight='bold')
plt.title(f"Custom-Built Gilbert G(n={n}, p={p}) Graph")
plt.show()
````

````python
#Solution Milestone 2, question c)
# random.seed(42)
# We create the graph
graph = directed_gilbert(10, 0.4)


def closeness(G):
    result = {}

    for s in G.nodes:
        # Distances from all nodes to u, d(v, u)
        # distance from s to every node (in-closeness)
        paths = nx.single_source_shortest_path_length(G.reverse(), s)
        sum_path_len = sum(paths.values())

        if sum_path_len > 0 and len(G) > 1:
          reachable_nodes = len(paths) - 1 # excluding itself
          # Wasserman - Faust formula, as used in NetworkX (normalized closeness)
          result[s] = ((len(paths) - 1) / sum_path_len)* (reachable_nodes / (len(G) - 1))
        else:
          result[s] = 0.0

    return result



manual_closeness = closeness(graph)

# We compute the Closeness Centrality with function
nx_closeness = nx.closeness_centrality(graph)

# We compare both results
print("Manual Closeness Centrality:\n", manual_closeness)
print("\nNetworkX Closeness Centrality:\n", nx_closeness)



````

d) Write a code that computes the Betweenness centrality for all nodes in the graph. *Note the code cannot use the function defined for this purpose in the NetworkX library*

Print the results of your solution along with the results obtained with the functions of NetworkX library to compute the Betweenness centrality. **Note tha both, your solution and NetworkX functions should provide the same results**

````python
#Solution Milestone 2, question d)

# We create the graph
G = directed_gilbert(10, 0.4)
n = G.number_of_nodes()

def betweenness_centrality_manual_directed(G, normalized=True, endpoints=False):
    betw = dict.fromkeys(G.nodes(), 0.0)

    for s in G.nodes():
        # Predecessors on shortest paths
        P = {v: [] for v in G.nodes()}
        # Number of shortest paths from s to v
        sigma = dict.fromkeys(G.nodes(), 0.0)
        sigma[s] = 1.0
        # Distance from s
        dist = dict.fromkeys(G.nodes(), -1)
        dist[s] = 0

        # Compute BFS on directed edges (successors)
        S = []
        Q = deque([s])
        while Q:
            v = Q.popleft()
            S.append(v)
            for w in G.successors(v):
                if dist[w] < 0:
                    dist[w] = dist[v] + 1
                    Q.append(w)
                if dist[w] == dist[v] + 1:
                    sigma[w] += sigma[v]
                    P[w].append(v)

        # Calculate the dependency accumulation
        delta = dict.fromkeys(G.nodes(), 0.0)
        while S:
            w = S.pop()
            for v in P[w]:
                if sigma[w] > 0.0:
                    delta[v] += (sigma[v] / sigma[w]) * (1.0 + delta[w])
            if w != s:
                betw[w] += delta[w]

    # We normalize dividing by (n-1)(n-2)
    if normalized and n > 2:
        scale = 1.0 / ((n - 1) * (n - 2))
        for v in betw:
            betw[v] *= scale

    return betw

# We manually compute the betweenness
manual_betweenness = betweenness_centrality_manual_directed(G, normalized=True, endpoints=False)

# We compute it with the NetworkX function
nx_betweenness = nx.betweenness_centrality(G, normalized=True, endpoints=False)

# We compare both results
print("Manual Betweenness Centrality (normalized):")
for k in sorted(manual_betweenness):
    print(f"  {k}: {manual_betweenness[k]:.6f}")

print("\nNetworkX Betweenness Centrality (normalized):")
for k in sorted(nx_betweenness):
    print(f"  {k}: {nx_betweenness[k]:.6f}")
````

d) Write a code that computes the Pagerank for all nodes in the graph. *Note the code cannot use the function defined for this purpose in the NetworkX library*
> Use a stop condition such that the sum of the Pagerank value of of all nodes in the graph for iteration i is < 10^-5 compared to iteration i-1.

Print the results of your solution along with the results obtained with the functions of NetworkX library to compute the Pagerank. **Note tha both, your solution and NetworkX functions should provide the same results.**



````python
#Solution Milestone 2, question e)

# We create the graph
G = directed_gilbert(10, 0.4)
n = G.number_of_nodes()
nodes = list(G.nodes())

alpha = 0.85
tol = 1e-5
max_iter = 1000

# Out-degree to spot dangling nodes
outdeg = {u: G.out_degree(u) for u in nodes}
dangling = [u for u in nodes if outdeg[u] == 0]

# We initialize PR uniformly
PR = {u: 1.0 / n for u in nodes}

# We power the iteration with dangling handling
for itr in range(1, max_iter + 1):
    newPR = {}
    # Total rank that dangles (no out-links)
    dangling_mass = alpha * sum(PR[u] for u in dangling) / n

    # Base term (random jump)
    base = (1.0 - alpha) / n

    # We distribute ranks
    for u in nodes:
        # Sum of contributions from predecessors
        contrib = 0.0
        for v in G.predecessors(u):
            if outdeg[v] > 0:
                contrib += PR[v] / outdeg[v]
        newPR[u] = base + dangling_mass + alpha * contrib

    # We compute the L1 distance between iterations
    diff = sum(abs(newPR[u] - PR[u]) for u in nodes)

    PR = newPR
    if diff < tol:
        break

# We compute it with the NetworkX function
nx_PR = nx.pagerank(G, alpha=alpha, tol=1e-12)

# We compare both results
print("Manual PageRank:")
for u in sorted(nodes):
    print(f"  {u}: {PR[u]:.8f}")

print("\nNetworkX PageRank:")
for u in sorted(nodes):
    print(f"  {u}: {nx_PR[u]:.8f}")
````

a) Write a code that  implements the k-core decomposition of a graph. The code should provide the k-core and the layer for each node in the graph. *Note the code cannot use the function defined for this purpose in the NetworkX library*

Print the results of your solution along with the results obtained with the functions of NetworkX library to compute the k-core (```core_number```) and layer (```onion_layers```). **Note that both, your solution and NetworkX functions should provide the same results**


````python
# We call the function created in 1b), directed_gilbert
random.seed(42)
G_dir = directed_gilbert(10, 0.4)

# As we are told to create an undirected graph, we have to convert it
G = G_dir.to_undirected()

import networkx as nx

def core_numbers_from_scratch(G: nx.Graph):
    H = G.copy()
    deg = dict(H.degree())
    core = {}
    removed = set()

    while len(removed) < H.number_of_nodes() + len(removed):  # We loop until all nodes are removed
        # We find a node not yet removed with the smallest current degree
        u = min((v for v in H.nodes if v not in removed), key=lambda v: deg[v])
        core[u] = deg[u]
        removed.add(u)
        # update neighbors' degrees
        for v in list(H.neighbors(u)):
            if v not in removed:
                deg[v] -= 1
        H.remove_node(u)
        # (deg for removed nodes is no longer used)

    return core



def onion_layers_from_scratch(G_undirected: nx.Graph):
    H = G_undirected.copy()
    layer_idx = 1
    onion = {}
    # Track current k-core threshold
    k = 1

    while H.number_of_nodes() > 0:
        # Find nodes with degree <= k
        candidates = [u for u, d in H.degree() if d <= k]
        if not candidates:
            k += 1
            continue
        # Remove the whole "batch" (the peel)
        for u in candidates:
            onion[u] = layer_idx
        H.remove_nodes_from(candidates)
        layer_idx += 1
    return onion


# 3) Run our implementations
my_core = core_numbers_from_scratch(G)
my_onion = onion_layers_from_scratch(G)

# 4) Compare with NetworkX (allowed here for validation/printing only)
nx_core = nx.core_number(G)
nx_onion = nx.onion_layers(G)

print("\n--- k-core numbers (ours vs NetworkX) ---")
for u in sorted(G.nodes()):
    print(f"Node {u}: ours={my_core[u]:2d} | nx={nx_core[u]:2d}")

print("\n--- Onion layers (ours vs NetworkX) ---")
for u in sorted(G.nodes()):
    print(f"Node {u}: ours={my_onion[u]:2d} | nx={nx_onion[u]:2d}")
````

````python
import networkx as nx
import time
from networkx.algorithms.community import label_propagation_communities, greedy_modularity_communities
from community import community_louvain
````

````python
#Solution Milestone 4, question a

# Predefined graphs
graphs = {
    "Karate Club": nx.karate_club_graph(),
    "Davis Southern Women": nx.davis_southern_women_graph(),
    "Florentine Families": nx.florentine_families_graph(),
    "Les Miserables": nx.les_miserables_graph()
}

# Function to print communities detected by each algorithm
def print_communities(algorithm_name, communities):
    print(f"{algorithm_name} Communities:")
    for i, community in enumerate(communities):
        print(f"Community {i + 1}: {community}")
    print("\n")

# Loop through each graph and apply the community detection algorithms
for graph_name, G in graphs.items():
    print(f"Processing {graph_name}...\n")

    # 1. Label Propagation Communities
    communities_lp = list(label_propagation_communities(G))
    print_communities("Label Propagation", communities_lp)

    # 2. Greedy Modularity Communities
    communities_gm = list(greedy_modularity_communities(G))
    print_communities("Greedy Modularity", communities_gm)

    # 3. Louvain Method
    partition = community_louvain.best_partition(G)

    # Grouping nodes by community ID
    communities_louvain = {}
    for node, community_id in partition.items():
        if community_id not in communities_louvain:
            communities_louvain[community_id] = []
        communities_louvain[community_id].append(node)

    # Convert dictionary to list of communities (list of lists)
    communities_louvain = list(communities_louvain.values())

    print_communities("Louvain", communities_louvain)
    print("-" * 50)  # Separator between graphs
````

````python
#Solution Milestone 4, question b
import time
from networkx.algorithms.community import label_propagation_communities, greedy_modularity_communities
from community import community_louvain

# List of predefined graphs
graphs = {
    "Karate Club": nx.karate_club_graph(),
    "Davis Southern Women": nx.davis_southern_women_graph(),
    "Florentine Families": nx.florentine_families_graph(),
    "Les Miserables": nx.les_miserables_graph()
}

# Results dictionary to store modularity and execution times
results = {}

# Loop through each graph
for graph_name, G in graphs.items():
    print(f"Processing {graph_name}...")

    # Store the results for the current graph
    results[graph_name] = {}

    # 1. Label Propagation Communities
    start_time = time.time()
    communities_lp = list(label_propagation_communities(G))
    time_lp = time.time() - start_time
    modularity_lp = nx.community.modularity(G, communities_lp)
    results[graph_name]["Label Propagation"] = (modularity_lp, time_lp)

    # 2. Greedy Modularity Communities
    start_time = time.time()
    communities_gm = list(greedy_modularity_communities(G))
    time_gm = time.time() - start_time
    modularity_gm = nx.community.modularity(G, communities_gm)
    results[graph_name]["Greedy Modularity"] = (modularity_gm, time_gm)

    # 3. Louvain Method
    start_time = time.time()
    partition = community_louvain.best_partition(G)
    communities_louvain = {}
    for node, community_id in partition.items():
        if community_id not in communities_louvain:
            communities_louvain[community_id] = []
        communities_louvain[community_id].append(node)
    communities_louvain = list(communities_louvain.values())
    time_louvain = time.time() - start_time
    modularity_louvain = nx.community.modularity(G, communities_louvain)
    results[graph_name]["Louvain"] = (modularity_louvain, time_louvain)

    print(f"Done with {graph_name}.\n")

# Print the results
for graph_name, algorithms in results.items():
    print(f"\nResults for {graph_name}:")
    for algorithm, (modularity, exec_time) in algorithms.items():
        print(f"{algorithm}: Modularity = {modularity:.4f}, Execution Time = {exec_time:.4f} seconds")
````

````python
#Solution Milestone 4, question c
# Analyzing the best algorithm based on modularity and execution time

# We will keep track of the best algorithm for each graph in terms of modularity and time
best_algorithms = {}

# Loop through the results and find the best algorithm for each graph
for graph_name, algorithms in results.items():
    best_modularity = -float('inf')
    best_time = float('inf')
    best_algorithm = None

    for algorithm, (modularity, exec_time) in algorithms.items():
        # Find the algorithm with the highest modularity and lowest time
        if modularity > best_modularity or (modularity == best_modularity and exec_time < best_time):
            best_modularity = modularity
            best_time = exec_time
            best_algorithm = algorithm

    best_algorithms[graph_name] = {
        "Best Algorithm": best_algorithm,
        "Modularity": best_modularity,
        "Execution Time": best_time
    }

# Print the best algorithm for each graph
print("\nBest Algorithm based on Modularity and Execution Time:")
for graph_name, result in best_algorithms.items():
    print(f"\n{graph_name}:")
    print(f"Best Algorithm: {result['Best Algorithm']}")
    print(f"Modularity: {result['Modularity']:.4f}")
    print(f"Execution Time: {result['Execution Time']:.4f} seconds")
````
